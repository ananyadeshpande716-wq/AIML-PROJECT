```python
from ultralytics import YOLO
import cv2
import matplotlib.pyplot as plt

# load model
model = YOLO("/content/drive/MyDrive/runs/detect/train/weights/best.pt")

# run prediction
results = model.predict(source="/content/Gemini_Generated_Image_ee6h6ee6h6ee6h6e.png", imgsz=1024, conf=0.25)

# get original image
img = cv2.imread("/content/Gemini_Generated_Image_ee6h6ee6h6ee6h6e.png")
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# draw boxes
for r in results:
    for box in r.boxes:
        cls_id = int(box.cls[0])
        name = model.names[cls_id]
        conf = float(box.conf[0])
        x1, y1, x2, y2 = map(int, box.xyxy[0])

        label = f"{name} {conf:.2f}"

        cv2.rectangle(img, (x1, y1), (x2, y2), (0, 255, 0), 2)
        cv2.putText(img, label, (x1, y1 - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

# show image
plt.imshow(img)
plt.axis("off")
plt.show()
```

```python
from ultralytics import YOLO

model = YOLO("yolov8n.pt")  # or best.pt if stable

model.train(
    data="/content/drive/MyDrive/unzipped_data/final_dataset1/data.yaml",

    epochs=120,
    imgsz=1024,
    batch=8,

    lr0=0.0005,
    lrf=0.01,

    momentum=0.937,
    weight_decay=0.0005,

    warmup_epochs=5,

    # 🔥 augmentations (VERY IMPORTANT)
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,

    degrees=5,
    translate=0.1,
    scale=0.5,
    shear=2,

    flipud=0.0,
    fliplr=0.5,

    mosaic=0.5,   # reduce from default (important for defects)
    mixup=0.1,

    # 🔥 helps small objects (cracks)
    close_mosaic=10,

    patience=30,
    workers=2
)
```
